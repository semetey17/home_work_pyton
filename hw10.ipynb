```python
import pandas as pd
```

```python
data = pd.read_csv('u.data', names = ['user id', 'item id', 'rating', 'timestamp'], sep = '\t', header=0)
data.head()
```

```python
max_rating = data['rating'].max()
max_rating_movies = data[ data.rating >= max_rating ]
max_rating_movies.groupby('item id').count().sort_values('user id', ascending=False).head(1)
```

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
filtered_countries = power[ (power['country']=='Lithuania') | (power['country']=='Latvia') | (power['country']=='Estonia')]
filtered_category = filtered_countries[ (filtered_countries['category']==4) | (filtered_countries['category']==12) | (filtered_countries['category']==21)]
filtered_year = filtered_category[ (filtered_category.year >= 2005) & (filtered_category.year <= 2010)]
filtered_quantity = filtered_year[ (filtered_year.quantity > 0)]
power_sum = filtered_quantity['quantity'].sum()
print(power_sum)
```
